```python
token_url = "http://gs25.gsretail.com/gscvs/ko/store-services/locations#;" # 해당 페이지 url
url = "http://gs25.gsretail.com/gscvs/ko/store-services/locationList?" # post url
with requests.Session() as s:
    token = s.get(token_url)
    token_text= BeautifulSoup(token.text).find("form", id="CSRFForm").find("input")['value']
    print(token_text)
    payload['CSRFToken'] = token_text
    r = s.post(url, data=payload, headers=head)
```

```python
from tqdm import tqdm
```

```python
data = []  # 데이터를 저장할 빈 리스트 생성

for page in tqdm(range(1, 3500)):
    payload['pageIndex'] = page  # 페이지 번호 업데이트
    r = requests.post(url, data=payload, headers=head)
    bs = BeautifulSoup(r.text, 'html.parser')

    for tr in bs.findAll("tr"):
        store_name = tr.find("span", class_="name").text if tr.find("span", class_="name") else ""
        store_tel = tr.find("span", class_="tel").text if tr.find("span", class_="tel") else ""
        detail_info_div = tr.find("div", class_="detail_info")
        store_address = detail_info_div.find("a").text if detail_info_div and detail_info_div.find("a") else ""
        if store_name or store_tel or store_address:  # 데이터가 있는 경우에만 추가
            data.append([store_name, store_tel, store_address])

# 리스트를 DataFrame으로 변환
df = pd.DataFrame(data, columns=["Store Name", "Telephone", "Address"])
```
