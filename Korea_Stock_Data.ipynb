```python
krx = "http://data.krx.co.kr/comm/bldAttendant/getJsonData.cmd"
```

```python
payload = {"bld": "dbms/MDC/STAT/standard/MDCSTAT01901",
"locale": "ko_KR",
"mktId": "ALL",
"share": "1",
"csvxls_isNo": "false",}
```

```python
krx_r = requests.post(krx, data=payload)
```

```python
df2 = pd.DataFrame(krx_r.json()['OutBlock_1'])
```

```python
df2
```

```python
filtered_df = df2[df2['MKT_TP_NM'].isin(['KOSDAQ', 'KOSPI'])]
```

```python
filtered_df
```

```python
stock_list = filtered_df['ISU_SRT_CD'].tolist()
```

```python
from tqdm import tqdm
```

```python
base_url = "https://api.finance.naver.com/siseJson.naver?symbol={}&requestType=2&count=2336&startTime=20240104&timeframe=day"
ma_list = [5, 20, 60, 90, 120]

for stock_code in tqdm(stock_list):
    url = base_url.format(stock_code)
    r = requests.get(url)
    df = pd.DataFrame(eval(r.text.strip())[1:], columns=eval(r.text.strip())[0])
    
    for ma in ma_list:
        df[f'ma{ma}'] = df['종가'].rolling(window=ma).mean()
    
    df.dropna(inplace=True)
    
    df['종목코드'] = "A" + stock_code

    file_name = f"{stock_code}.csv"
    df.to_csv(f"./stock/{stock_code}.csv", index=False, encoding='utf-8')
```
