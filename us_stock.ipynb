# 종목 코드 

```python
url = "https://www.hankyung.com/globalmarket/usa-marketcap-nyse"
requests.get(url).text
bs = BS(requests.get(url).text)
x = bs.findAll("div", class_="symbol txt-en")
stock_code = [element.text for element in x]
```

```python
stock_code
```

# 데이터 프레임 만들기

```python
import pandas as pd

# 리스트를 튜플로 결합
combined_list = list(zip(stock_code, stock_names))

# 튜플의 리스트를 DataFrame으로 변환
df = pd.DataFrame(combined_list, columns=['종목코드', '종목이름'])
```

# 셀리니움 적용

```python
from tqdm import tqdm
```

```python
url = "https://www.hankyung.com/globalmarket/usa-marketcap-nyse?page={}"

for i in tqdm(range(1, 42)):
    for_1 = url.format(i)
    
    response = requests.get(for_1)
    soup = BS(response.text, 'html.parser')
    
    x = soup.findAll("div", class_="symbol txt-en")
    stock_codes = [element.text for element in x]

    y = soup.findAll("div", class_="stock-name")
    stock_names = [element.text for element in y]
    
    combined_list.extend(list(zip(stock_codes, stock_names)))

df = pd.DataFrame(combined_list, columns=['종목코드', '종목이름'])
```

```python
df.to_csv(f"./US_STOCK/us_stock.csv", index=False, encoding='utf-8')
```
