```python
import requests 
url = "https://finance.naver.com/marketindex/exchangeDailyQuote.naver?marketindexCd=FX_USDKRW&"
usd = pd.concat([pd.read_html(requests.get(url.format(x)).text)[0]])

usd.columns = [x[0] for x in usd.columns]
usd_df = usd[['날짜', '매매기준율']].copy()
usd_df.날짜 = pd.to_datetime(usd_df.날짜)
```

```python
filtered_row = usd_df[usd_df['날짜'] == '2024-01-02']
```

```python
from tqdm import tqdm
```

```python
ex_240102_rate = float(exchange_rate_20230102)
```

```python
ex_240102_rate
```

```python

for player_id in tqdm(playid):
    # 각 선수의 프로필 페이지로 이동
    url = f"https://www.koreabaseball.com/Record/Player/PitcherDetail/Basic.aspx?playerId={player_id}"
    driver.get(url)
    
    # 선수 데이터 추출
    player_data = driver.find_element(By.CSS_SELECTOR, "#contents > div.sub-content > div.player_info > div.player_basic").text
    
    player_dict = {}
    for line in player_data.split('\n'):
        if ': ' in line:
            key, value = line.split(': ', 1)
            player_dict[key] = value
    
    # '연봉' 정보가 있는지 확인하고 처리
    if '연봉' in player_dict:
        salary_value = player_dict['연봉']
        
        # "달러" 문자열이 포함되어 있는 경우 처리
        if '달러' in salary_value:
            # 환율 적용 (ex_240102_rate는 예시 환율입니다. 실제 환율을 사용하십시오.)
            ex_240102_rate = 1311  # 예시 환율
            salary_value = int(salary_value.replace('달러', '').replace(',', '')) * ex_240102_rate
            player_dict['연봉'] = f"{salary_value // 1}"
    
    player_df = pd.DataFrame([player_dict])
    if '연봉' not in player_df.columns:
        player_df['연봉'] = '0'
    else:
        player_df['연봉'] = player_df['연봉'].fillna('0')

    player_df.to_csv(f"./kbo/{player_id}.csv", index=False, encoding='utf-8')
```
